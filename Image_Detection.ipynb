Importing Libraries

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

Loading the
YOLO Model

```python
print("LOADING YOLO")
net = cv2.dnn.readNet("Image_detection/yolov3.weights", "Image_detection/yolov3.cfg")
```

Loading
COCO
Class
Names

```python
classes = []
with open("Image_detection/coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
```

Getting Layer
Names of
the Network

```python
layer_names = net.getLayerNames()
```

Determine
the output
layer
names from
the YOLO
model

```python
output_layers = [layer_names[i- 1] for i in net.getUnconnectedOutLayers()]
print("YOLO LOADED")
```

Reading and
Preprocessing
the Image

```python
img = cv2.imread("Image_detection/person.jpg")
img = cv2.resize(img, None,fx=0.4, fy=0.4)
height, width, channels =img.shape
print("Input Image:")
cv2_imshow(img)
```

Blob Creation
and
Preprocessing

```python
blob =cv2.dnn.blobFromImage(img, 1 /255.0, (416, 416),swapRB=True,crop=False)
```

Object
Detection

```python
net.setInput(blob)
outs = net.forward(output_layers)
```

Processing
YOLO
Outputs

```python
class_ids = []
confidences = []
boxes = []
for out in outs:
  for detection in out:
    scores = detection[5:]
    class_id = np.argmax(scores)
    confidence = scores[class_id]
    if confidence > 0.3:
      # Object detected
      center_x = int(detection[0] * width)
      center_y = int(detection[1] * height)
      w = int(detection[2] * width)
      h = int(detection[3] * height)
      # Rectangle coordinates
      x = int(center_x - w / 2)
      y = int(center_y - h / 2)
      boxes.append([x, y, w, h])
      confidences.append(float(confidence))
      class_ids.append(class_id)
```

Non-
Maximum
Suppression
(NMS)

```python
indexes =cv2.dnn.NMSBoxes(boxes,confidences, 0.5, 0.4)
```

Drawing
Bounding
Boxes and
Labels

```python
colors = np.random.uniform(0, 255,size=(len(classes), 3))
for i in range(len(boxes)):
    if i in indexes:
      x, y, w, h = boxes[i]
      label =str(classes[class_ids[i]])
      color = colors[class_ids[i]]
```

```python
# Draw the bounding box
cv2.rectangle(img, (x, y), (x +
w, y + h), color, 2)
# Draw the label
cv2.putText(img, label, (x, y -
5), cv2.FONT_HERSHEY_SIMPLEX, 1/2,
color, 2)
print(f"Detected object:{label}, confidence:{confidences[i]}, box: {x}, {y},{w}, {h}")
```

Displaying
the Result

```python
# Display the image
cv2_imshow(img)
cv2.waitKey(0)
```
